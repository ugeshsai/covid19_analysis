```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data_url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(data_url)
```

```python
country = 'India'
country_df = df[df['location'] == country].copy()
```

```python
country_df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(country_df['date'], country_df['new_cases'], marker='o')
plt.title(f'Daily New COVID-19 Cases in {country}')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
fig = px.line(country_df, x='date', y=['total_cases', 'total_deaths'], 
              title=f'Total COVID-19 Cases and Deaths in {country}')
fig.update_layout(xaxis_title='Date', yaxis_title='Count')
fig.show()
```

```python
country_df['recovery_rate'] = country_df['total_cases'] / country_df['total_cases'].shift(1)
country_df['fatality_rate'] = country_df['total_deaths'] / country_df['total_cases'] * 100
```

```python
plt.figure(figsize=(12, 6))
plt.plot(country_df['date'], country_df['recovery_rate'], marker='o', label='Recovery Rate')
plt.plot(country_df['date'], country_df['fatality_rate'], marker='o', label='Fatality Rate')
plt.title(f'Recovery Rate and Fatality Rate in {country}')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
country_df['testing_positivity_rate'] = (country_df['new_cases'] / country_df['new_tests']) * 100
```

```python
plt.figure(figsize=(12, 6))
plt.plot(country_df['date'], country_df['testing_positivity_rate'], marker='o', color='purple')
plt.title(f'Testing Positivity Rate in {country}')
plt.xlabel('Date')
plt.ylabel('Positivity Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
